```python
import numpy as np

# Lista de tiendas a comparar
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
top_n = 5  # Número de categorías principales

# Preparar datos: obtener las top N categorías para cada tienda y construir DataFrame
df = pd.DataFrame({
    tienda: ventas_por_categoria[tienda].head(top_n)
    for tienda in tiendas
}).fillna(0)

# Configurar gráficos
fig, ax = plt.subplots(figsize=(12, 6))
indices = np.arange(top_n)
width = 0.2

# Color en tonos turquesa claros
colors = ['#81d4fa', '#4fc3f7', '#29b6f6', '#03a9f4']

# Graficar barras agrupadas
for i, tienda in enumerate(tiendas):
    ax.bar(indices + i * width, df[tienda], width=width, label=tienda, color=colors[i])

# Configurar eje x
ax.set_xticks(indices + width * (len(tiendas) - 1) / 2)
ax.set_xticklabels(df.index)
ax.set_xlabel('Categorías')
ax.set_ylabel('Ventas')
ax.set_title('Comparativa de Principales Categorías por Tienda')
ax.legend()

plt.tight_layout()
plt.show()

```

###**Análisis y Recomendación**

**1. Introducción**

En el presente informe se analizaron los datos de ventas, costos y satisfacción del cliente de cuatro sucursales (Tienda 1, Tienda 2, Tienda 3 y Tienda 4) de AluraStore. El propósito final es proporcionar una recomendación fundamentada al Sr. Juan sobre cuál de estas tiendas debería vender para iniciar un nuevo emprendimiento.

La decisión se basará en los análisis que buscan retener las sucursales con mejor potencial a futuro.

**2. Desarrollo y Análisis de Datos**

Para la recomendación, se tuvieron en cuenta los siguientes factores que surgen del análisis:

***2.1. Ingresos Totales y Costos de Envío***

Los ingresos totales son un indicador del tamaño del mercado de cada tienda, pero deben considerarse junto con los costos de envío que impactan su rentabilidad operativa.

Ingresos Totales: La Tienda 1 es la que genera mayores ingresos (aprox.  1,150M), seguida de cerca por las Tiendas 2 y 3, siendo la Tienda 4 la que presenta los ingresos más bajos (1,038M).

Costo de Envío Promedio: aquí la relación es inversa. La Tienda 4, con los menores ingresos, tiene también el costo de envío promedio más bajo (aprox.  23,459). Por el contrario, la Tienda 1, con los mayores ingresos, presenta el costo de envío promedio más alto (aprox. 26,018).

***2.2. Análisis de Categorías y Productos***

El análisis de categorías nos indica dónde se concentra la demanda  de los clientes y el valor de cada tienda.

Categorías Más Vendidas (por Ingresos): En todas las tiendas, "Electrónicos" es la categoría que genera más dinero, seguida por "Electrodomésticos" y "Muebles".

Categorías Más Vendidas (por Cantidad): En todas las tiendas, "Muebles" y "Electrónicos" son las categorías con el mayor número de ventas individuales.

Productos Individuales: Los productos específicos más vendidos (ej. "Armario", "Cama box") varían significativamente entre tiendas, lo que sugiere que el éxito de un producto individual puede depender de la tienda, pero las categorías de alto valor (Electrónicos) son fuertes en todas.

***2.3. Calificación Promedio de Clientes***

La satisfacción del cliente es un indicador vital de la salud y el éxito a largo plazo de una sucursal.

Calculamos el promedio ponderado para obtener una métrica comparable:

Tienda 3: 4.04 (La más alta)
Tienda 2: 4.03
Tienda 4: 3.99
Tienda 1: 3.98 (La más baja)

Se observa que, aunque las diferencias son pequeñas, las Tiendas 3 y 2 lideran en satisfacción del cliente. La Tienda 1, a pesar de ser la de mayores ingresos, tiene la calificación promedio más baja.
  
**3. Conclusión y Recomendación**

Tras analizar todos los factores para determinar qué sucursal debería vender el Sr. Juan para financiar un nuevo emprendimiento, se determina lo siguiente:

**Tienda 4**: Es la que genera menos ingresos (1,038M). Venderla aportaría el menor capital, aunque también es el mercado más pequeño.

**Tienda 3**: Es la tienda más saludable. Tiene la calificación de cliente más alta (4.04) y costos de envío muy bajos (24,805). Sería la peor tienda para vender, ya que es la más prometedora para conservar.

**Tienda 2**: Similar a la Tienda 3, es muy saludable, con la segunda calificación más alta (4.03) e ingresos sólidos (1,116M).

**Tienda 1**: Presenta un perfil problemático. Aunque puede ser atractiva por generar los ingresos más altos (1,150M), también sufre de los costos de envío promedio más altos (26,018) y, lo más importante, tiene la calificación de cliente promedio más baja (3.98).

**Recomendación Final:** Se recomienda al Sr. Juan que elija la Tienda 1 para venderla.

**Justificación:**
Si bien la Tienda 1 genera los mayores ingresos (lo que podría darle un buen precio de venta), es estratégicamente la sucursal más débil para mantener a largo plazo:

***Peor Satisfacción del Cliente***: Con la calificación promedio más baja (3.98), la Tienda 1 tiene los clientes más insatisfechos. Esto representa un riesgo de reputación y una mayor dificultad para retener clientes en el futuro.

***Costos Operativos Más Altos***: Sus costos de envío son los más caros (26,018). Vender esta sucursal elimina la operación más costosa de la red.

Vender la Tienda 1 le permite al Sr. Juan obtener un buen capital (dado sus altos ingresos) mientras se deshace de la sucursal con los peores indicadores de sostenibilidad (altos costos y baja satisfacción), permitiéndole conservar las tiendas 2 y 3, que son más eficientes y queridas por los clientes.
