# 4. Productos más y menos vendidos

```python
productos_t1 = display(tienda.groupby('Producto').size().sort_values(ascending=False))
```

```python
import matplotlib.pyplot as plt
import pandas as pd

productos_vendidos = tienda.groupby('Producto').size().sort_values(ascending=False)
mas_vendidos = productos_vendidos.max()
menos_vendidos = productos_vendidos.min()
productos_mas_vendidos = productos_vendidos[productos_vendidos == mas_vendidos]
productos_menos_vendidos = productos_vendidos[productos_vendidos == menos_vendidos]

print("Productos más vendidos en Tienda 1:")
display(productos_mas_vendidos)

print("\nProductos menos vendidos en Tienda 1:")
display(productos_menos_vendidos)

productos_grafico = pd.concat([productos_mas_vendidos, productos_menos_vendidos])

colors = ['purple'] * len(productos_mas_vendidos) + ['yellow'] * len(productos_menos_vendidos)

plt.figure(figsize=(8, 4))
productos_grafico.plot(kind='barh', color=colors)
plt.title('Productos Más y Menos Vendidos en Tienda 1')
plt.xlabel('Cantidad Vendida')
plt.ylabel('Producto')
plt.tight_layout()
display(plt.show())
```

```python
import matplotlib.pyplot as plt

costos_envio_promedio = {
    'Tienda 1': round(tienda['Costo de envío'].mean(), 2),
    'Tienda 2': round(tienda2['Costo de envío'].mean(), 2),
    'Tienda 3': round(tienda3['Costo de envío'].mean(), 2),
    'Tienda 4': round(tienda4['Costo de envío'].mean(), 2)
}

colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.figure(figsize=(4, 4))
plt.bar(costos_envio_promedio.keys(), costos_envio_promedio.values(), color=colors)
plt.title('Costo de Envío Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de Envío Promedio')
plt.show()
```

## Ingresos por tienda

### Subtask:
Crear un gráfico de barras que muestre la suma total de ingresos (`Precio`) para cada una de las cuatro tiendas.


**Reasoning**:
Create a dictionary with total revenue for each store and then generate a bar plot to visualize the total revenue per store.



```python
import matplotlib.pyplot as plt

ingresos_totales = {
    'Tienda 1': tienda['Precio'].sum(),
    'Tienda 2': tienda2['Precio'].sum(),
    'Tienda 3': tienda3['Precio'].sum(),
    'Tienda 4': tienda4['Precio'].sum()
}

plt.figure(figsize=(8, 6))
plt.bar(ingresos_totales.keys(), ingresos_totales.values(), color=['blue', 'green', 'red', 'purple'])
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos')
plt.show()
```

## Distribución de categorías de productos

### Subtask:
Generar gráficos de barras para cada tienda, mostrando la cantidad de productos por `Categoría del Producto`.


**Reasoning**:
Group each store's DataFrame by 'Categoría del Producto', count the occurrences, and then create a bar plot for each to visualize the distribution of products by category.



```python
plt.figure(figsize=(10, 6))
tienda.groupby('Categoría del Producto').size().sort_values(ascending=False).plot(kind='bar')
plt.title('Distribución de Categorías de Productos en Tienda 1')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Productos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False).plot(kind='bar', color='green')
plt.title('Distribución de Categorías de Productos en Tienda 2')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Productos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False).plot(kind='bar', color='red')
plt.title('Distribución de Categorías de Productos en Tienda 3')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Productos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False).plot(kind='bar', color='purple')
plt.title('Distribución de Categorías de Productos en Tienda 4')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Productos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
