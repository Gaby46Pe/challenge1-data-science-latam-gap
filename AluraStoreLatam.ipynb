# 4. Productos más y menos vendidos

```python
productos_t1 = display(tienda.groupby('Producto').size().sort_values(ascending=False))
```

```python
import matplotlib.pyplot as plt
import pandas as pd

productos_vendidos = tienda.groupby('Producto').size().sort_values(ascending=False)
mas_vendidos = productos_vendidos.max()
menos_vendidos = productos_vendidos.min()
productos_mas_vendidos = productos_vendidos[productos_vendidos == mas_vendidos]
productos_menos_vendidos = productos_vendidos[productos_vendidos == menos_vendidos]

print("Productos más vendidos en Tienda 1:")
display(productos_mas_vendidos)

print("\nProductos menos vendidos en Tienda 1:")
display(productos_menos_vendidos)

# Combine the most and least sold products for plotting
productos_grafico = pd.concat([productos_mas_vendidos, productos_menos_vendidos])

# Create a color list to highlight most and least sold
colors = ['purple'] * len(productos_mas_vendidos) + ['yellow'] * len(productos_menos_vendidos)

# Create the bar plot
plt.figure(figsize=(8, 4))
productos_grafico.plot(kind='barh', color=colors)
plt.title('Productos Más y Menos Vendidos en Tienda 1')
plt.xlabel('Cantidad Vendida')
plt.ylabel('Producto')
plt.tight_layout()
display(plt.show())
```
